```python
import pandas as pd
import plotly.express as px
file_path = 'C:/Users/Vinay/Downloads/archive/cleaned_dataset/metadata.csv'  
data = pd.read_csv(file_path)
data.head(20)

```

```python
data.tail(20)
```

```python
impedance_data = data[data['type'] == 'impedance']
impedance_data = impedance_data.sort_values(by='test_id')
impedance_data['Aging Metric'] = impedance_data['test_id']
impedance_data['Re'] = pd.to_numeric(impedance_data['Re'], errors='coerce')
impedance_data['Rct'] = pd.to_numeric(impedance_data['Rct'], errors='coerce')
fig_re = px.line(
    impedance_data, 
    x='Aging Metric', 
    y='Re', 
    title='Change in Electrolyte Resistance (Re) Over Aging',
    labels={'Aging Metric': 'Aging Metric (Test ID)', 'Re': 'Electrolyte Resistance (Ohms)'},
    markers=True
)
fig_re.show()
```

```python
fig_rct = px.line(
    impedance_data, 
    x='Aging Metric', 
    y='Rct', 
    title='Change in Charge Transfer Resistance (Rct) Over Aging',
    labels={'Aging Metric': 'Aging Metric (Test ID)', 'Rct': 'Charge Transfer Resistance (Ohms)'},
    markers=True
)
fig_rct.show()
```
